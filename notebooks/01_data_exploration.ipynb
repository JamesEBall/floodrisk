# 01 — CAMELS Data Exploration

This notebook explores the CAMELS-US dataset: basin forcing variables, streamflow
observations, and static catchment attributes.

```python
from floodrisk.data.camels import CAMELSLoader

DATA_DIR = "../data/camels"  # adjust to your download path
loader = CAMELSLoader(DATA_DIR)
```

```python
basins = loader.list_basins()
print(f"Number of basins: {len(basins)}")
print(f"First 10 basins: {basins[:10]}")
```

```python
# Load forcing and streamflow for a sample basin
basin_id = basins[0]
forcing = loader.load_basin_forcing(basin_id)
streamflow = loader.load_basin_streamflow(basin_id)

print(f"Forcing shape: {forcing.shape}")
print(f"Forcing columns: {list(forcing.columns)}")
print(f"Streamflow length: {len(streamflow)}")
forcing.head()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
axes[0].plot(forcing.index, forcing["prcp"], label="Precipitation")
axes[0].set_ylabel("Precip (mm/day)")
axes[0].legend()
axes[1].plot(streamflow.index, streamflow.values, label="Streamflow", color="tab:blue")
axes[1].set_ylabel("Discharge (ft³/s)")
axes[1].legend()
plt.suptitle(f"Basin {basin_id}")
plt.tight_layout()
plt.show()
```

```python
# Static catchment attributes
attrs = loader.get_basin_attributes(basin_id)
print(f"Number of attributes: {len(attrs)}")
attrs.head(20)
```
